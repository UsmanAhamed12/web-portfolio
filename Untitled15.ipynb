```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Employee Dataset.csv')
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x = 'age',y = 'salary',data = data)

plt.title('Scatter plot age vs salary')
plt.xlabel('age')
plt.ylabel('salary')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x = 'healthy_eating',y = 'active_lifestyle',data = data)

plt.title('Scattter plot healthy eating vs active lifestyle')
plt.xlabel('healthy eating')
plt.ylabel('active lifestyle')
plt.hlines(xmax = 9,xmin = 0, y =6)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x = 'healthy_eating',y = 'salary',data = data)

plt.title('Scatter plot healthy eating vs salary')
plt.xlabel('healthy eating')
plt.ylabel('salary')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(x = 'groups',data = data, color = 'red', edgecolor = 'linen',alpha = 0.5, bins = 5)

plt.title('Histogram plot for the group')
plt.xlabel('group')
plt.ylabel('count')

plt.show()
```

```python
plt.figure(figsize = (8,4))

sns.histplot(x = 'salary',data = data,bins = 10,alpha = 0.7,color = 'orange',edgecolor = 'linen')

plt.title('Count plot for the salary')
plt.xlabel('Employee salary')
plt.ylabel('Count')

plt.show()
```
